```python
import numpy as np
import matplotlib.pyplot as plt

class ChessBoard: 
    
    def __init__(self): 
        #establish color values
        self.black = (0,0,0)
        self.white = (1,1,1)
        self.blue = (0,1,1)
        self.red =  (1,.2,0)
        self.red_pieces = {'queen': (0,0)}
        self.blue_pieces = {'queen': (0,0)}
        
        #creates board with horizontal, vertical and cell color values, default value of white
        self.chessboard = np.ones((8,8,3))
        
        #create board with alternating cell colors
        for row in range(8): 
            for col in range(8): 
                if (row + col) % 2 == 0:
                    self.chessboard[row, col] = self.white
                else: 
                    self.chessboard[row, col] = self.black
    
    #renders the chessboard
    def render(self): 
        plt.imshow(self.chessboard)
        plt.show()


        
    def add_red(self, row, col): 
        self.red_pieces['queen'] = (row, col)
        self.chessboard[row][col] = self.red
    
    
    def add_blue(self,row, col): 
        self.blue_pieces['queen'] = (row, col)
        self.chessboard[row][col] = self.blue
    
    
    def is_under_attack(self):
        #find position of blue and red pieces
        blue_col, blue_row = self.blue_pieces['queen']
        red_col, red_row = self.red_pieces['queen']

    # test col and row for match
        if red_col == blue_col:
            return True
        if red_row == blue_row:
            return True

    # test diagonals for match by using the col and row values to take the absolute value of the result
        row_diff = abs(red_col - blue_col)
        col_diff = abs(red_row - blue_row)
        if  row_diff == col_diff:
            return True

        return False

board = ChessBoard()

board.add_red(0,0)
board.add_blue(5,6)
board.render()
print(board.is_under_attack())
```
